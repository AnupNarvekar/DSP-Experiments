```
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

**Unit Step Signal**

```
## Unit Step Signal

for i in range (-5,6):
  if i >= 0:
    a = 1
    plt.plot(i, 1, 'o:k')
    plt.vlines(i,0,1)
  else:
    a = 0
    plt.plot(i,a, 'o:k')

plt.title('Unit Step Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid()
plt.show()
```

**Impulse / Dirac Delta Signal**

```
## Impulse / Dirac Delta Signal

for i in range (-5,6):
  if i == 0:
    a=1
    plt.plot(i,1, 'o:k')
    plt.vlines(i,0,1)
  else :
    a=0
    plt.plot(i,a, 'o:k')


plt.title('Impulse Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid()
plt.show()
```

**Ramp Signal**

```
## Ramp Signal

for i in range(-9, 11):
  if i >= 0 :
    a=i
    plt.plot(i,a, 'o:k')
    plt.vlines(i,0,i)


plt.title('Ramp Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid()
plt.show()
```

**Exponential signals**

```
### Decaying Exponentially

import math

a  = float(input('Enter a floating point number: '))    # a = 0.5
for n in range (-5,6):
  x=pow(a,n)
  plt.plot(n,x, 'o:k')
  plt.vlines(n,0,x)

plt.title('Exponential Signals')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid()
plt.show()
```
